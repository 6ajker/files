```python
Необходимо реализовать пользовательские команды (внимание! это не название функций, \
    которые должны быть выразительными, а команды, которые вводит пользователь,\
    чтобы получить необходимый результат):
  
- p – people – команда, которая спросит номер документа и выведет имя человека, которому\
    он принадлежит;  
- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой\
    он находится  
*Правильно обработайте ситуации, когда пользователь будет вводить\
    несуществующий документ*.
- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
- as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень.\
    *Предусмотрите случай, когда пользователь добавляет полку, которая уже существует*.
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из\
    перечня полок. *Предусмотрите сценарий, когда пользователь вводит несуществующий \
    документ*;  
- m – move – команда, которая спросит номер документа и целевую полку и переместит его\
    с текущей полки на целевую. *Корректно обработайте кейсы, когда пользователь пытается\
    переместить несуществующий документ или переместить документ на несуществующую полку*;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив\
    его номер, тип, имя владельца и номер полки, на котором он будет храниться.\
    *Корректно обработайте ситуацию, когда пользователь будет пытаться добавить \
    документ на несуществующую полку*.
```

```python

# Каталог документов:
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

# Перечень полок, на которых находятся документы:
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}


def init():
    while True:
        print()
        enter = input("Введите 'var'для просмотра возможных запросов, 'exit' для выхода или введите необходимый запрос\n")
        if enter == "exit":
            break
        elif enter == "var":
            var()
        elif enter == "p":
            people()
        elif enter == "s":
            shelf()
        elif enter == "l":
            list_all()
        elif enter == "as":
            add_shelf()
        elif enter == "d":
            delete_num()
        elif enter == "m":
            move()
        elif enter == "a":
            add()
        else:
            print("Неверный ввод")


def var():
    """команда, которая выводит возможные варианты запросов"""
    print("Возможные запросы:\n"
          "exit - выход\n"
          "p - запрос имени человека по указанному номеру документа\n"
          "s - запрос номера полки по указанному номеру документа\n"
          "l - вывод отформатированного списка всех документов\n"
          "as - добавление указанного номера полки в перечень\n"
          "d - удаление указанного номера документа из каталога и с полок\n"
          "m - перемещение указанного документа на указанную полку\n"
          "a - создание нового документа(с вводом данных) на указанной полке\n")


def people():
    """команда, которая спросит номер документа и выведет имя человека,
    которому он принадлежит"""
    number = input("Введите номер документа\n")
    x = 0
    for i in documents:
        if number in i.values():
            print(i['name'])
            x += 1
    if x == 0:
        print("Указанный номер документа не найден")


def shelf():
    """команда, которая спросит номер документа и выведет номер полки,
    на которой он находится"""
    shelf_x = input("Введите номер документа\n")
    x = 0
    for i in directories.values():
        for j in i:
            if shelf_x == j:
                for key in directories:
                    if directories[key] == i:
                        print(key)
                        x +=1
    if x == 0:
        print("Указанный номер документа не найден")


def list_all():
    """команда, которая выведет список всех документов в формате
    **passport "2207 876234" "Василий Гупкин"**"""
    for i in documents:

        print(f"**{i['type']} \"{i['number']}\" \"{i['name']}\"**")


def add_shelf():
    """команда, которая спросит номер новой полки и добавит ее в перечень"""
    new = input("Введите номер новой полки\n")
    x = 0
    for key in directories:
        if key == new:
            print("Такая полка уже существует")
            x +=1

    if not new.isdecimal():
        print("Номер новой полки необходимо вводить цифрами")

    if new.isdecimal()and x == 0:
        directories[new] = []
        print(f"Полка с номером {new} успешно добавлена")

    print(directories)


def delete_num():
    """команда, которая спросит номер документа и удалит его из каталога
    и из перечня полок"""
    x = 0
    kill = input("Введите номер удаляемого документа\n")
    for i in directories.values():
        for j in i:
            if kill == j:
                print("Указанный документ успешно удален")
                x +=1
                i.remove(j) #удаляем с полки

                for z in documents:
                    if kill in z.values():
                        documents.remove(z)  #удаляем из документов
                print("Полки:\n", directories)
                print("Документы:\n", documents)
    if x == 0:
        print("Указанный документ не найден")


def move():
    """команда, которая спросит номер документа и целевую полку и переместит его
    с текущей полки на целевую"""
    x = 0
    y = 0
    move_num = input("Введите номер перемещаемого документа\n")
    were = input("Введите номер желаемой полки\n")
    for i in directories.values():
        for j in i:
            if move_num == j:
                x += 1
                for key in directories:
                    if key == were:
                        y +=1
                        directories[key].append(move_num)
                        i.remove(j)

    if y == 0:
        print("Указанная полка не найдена")

    if x == 0:
        print("Документа с указанным номером не найдено")
    print("Полки:\n", directories)


def add():
    """команда, которая добавит новый документ в каталог и в перечень полок,
    спросив его номер, тип, имя владельца и номер полки,
    на котором он будет храниться"""
    new = {}
    type_new = input("Введите тип документа\n")
    number_new = input("Введите номер документа\n")
    name_new = input("Введите имя владельца\n")
    shelf_new = input("Введите номер полки\n")

    x = 0
    for key in directories:
        if key == shelf_new:
            x += 1
    while x == 0:
        print("Вы указали не существующую полку")
        shelf_new = input("Введите номер полки\n")
        for key in directories:
            if key == shelf_new:
                x += 1
    new['type'] = type_new
    new['number'] = number_new
    new['name'] = name_new
    documents.append(new)

    for key in directories:
        if key == shelf_new:
            directories[key].append(new['number'])

    print("Новый документ: ", new)
    print("Полки: \n", directories)
    print("Документы: \n", documents)


init()
```
